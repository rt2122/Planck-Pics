```python
import numpy as np
import pandas as pd
import healpy as hp
from modules.DS_healpix_fragmentation import radec2pix
from modules.DS_data_transformation import get_cat_name
from modules.DS_models_stat import cut_cat
%config InlineBackend.figure_format = 'retina'
```

```python
def pic_df(filename, name=None, nside=2**7, dict_cut={}):
    df = cut_cat(pd.read_csv(filename), dict_cut=dict_cut)
    if name is None:
        name = get_cat_name(filename)
    
    a = np.zeros(hp.nside2npix(nside))
    a[radec2pix(df['RA'], df['DEC'], nside)] = 1
    hp.mollview(a, nest=True, cbar=False, title=name)
```

```python
pic_df('./Data/original_cats/PSZ2.csv')
```
